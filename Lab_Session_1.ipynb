**Lab Task 1**: Plot all the given functions to observe the roots by visualization, fill the table by your visual guess of root. We have plotted one function for you.</br>
1) $f(x)=cos(x)-1.3x$</br>
2) $f(x)=xcos(x)-2x^2+3x-1$</br>
3) $f(x)=2xcos(2x)-(x+1)^2$





```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10, 10]

def f(x):
    return np.cos(x)-1.3*x

x = np.linspace(-10, 10, 500)
y = f(x)

plt.plot(x, y, color='red')
plt.hlines(y=0, xmin=-10, xmax=10, color='blue')


roots_x = []
roots_y = []

for i in range(len(x) - 1):
    if np.sign(y[i]) != np.sign(y[i+1]):
        root_x = (x[i] + x[i+1]) / 2
        root_y = f(root_x)
        roots_x.append(root_x)
        roots_y.append(root_y)
        plt.scatter(root_x, root_y, color='green', label='Root')
plt.xlim(-root_x, root_x)
plt.legend()
plt.show()

print("Approximate roots found thorugh visualization:", roots_x)
```

**Lab Task 5**: Write program of Secant and False Position method by altering above codes.

**Secant Method**</br>f(x)=cos x -1.3x

```python
import math

def f(x):
    return math.cos(x) - 1.3*x

def secant_method(x0, x1, tol, max_iter):
    iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        x_next = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if abs(x_next - x1) < tol:
            return x_next
        x0, x1 = x1, x_next
    print("No root found.")
    return None

x0 = 0.0
x1 = 1.0

tolerance = 1e-6
max_iterations = 1000

root = secant_method(x0, x1, tolerance, max_iterations)

if root is not None:
    print("Approximated root:", root)
```

**False Position Method**</br>f(x)=cos x -1.3x

```python
import math

def f(x):
    return math.cos(x) - 1.3*x

def false_position_method(a, b, tol, max_iter):

    iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        if f(c) == 0 or abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    print("No root found")
    return None

a = 0.0
b = 1.0

tolerance = 1e-6
max_iterations = 1000

root = false_position_method(a, b, tolerance, max_iterations)

if root is not None:
    print("Approximated root:", root)
```
